Tìm cột bị thiếu

```python
for item in track_df.columns:
    check = track_df[item].isnull()
    if check.sum() > 0:
        print("Column which missing data: {}".format(item))
```

```python
def DelMissCol(df,percent):
    list_col = df.columns
    for col in list_col:
        check = (df[col].isnull()) #Tạo list check để lưu các giá trị null của cột(True là Null, False là khác Null)
        if(check.sum() > len(df)*percent): #Kiểm tra số giá trị rỗng có hơn số phần trăm cho trước
            del df[col]
    return df
track_df = DelMissCol(track_df,0.5)
```

```python
def DelMissRow(df,percent):
    dft = df.transpose()  #Chuyển vị để dễ thao tác với hàng hơn
    del_lis = []
    for item in dft.columns:
        check = (dft[item].isnull()) #Tạo list check để lưu các giá trị null của cột(True là Null, False là khác Null)
        if(check.sum() > len(dft)*percent): #Kiểm tra số giá trị rỗng có hơn số phần trăm cho trước
            del_lis.append(int(item))
    df = df.drop(df.index[del_lis])
    return df

track_df = DelMissRow(track_df,0.5)
```

```python
list_date = ['created_at','last_modified','display_date']
def Format_day(df, list_date):
    format_ = '%Y-%m-%dT%H:%M:%SZ'
    n = len(track_df['created_at'])
    for item in list_date:
        check = track_df[item].isnull()
        for i in range(n):
            if check[i] == False:
                track_df[item][i] = datetime.datetime.strptime(track_df[item][i],format_)
    return df
track_df = Format_day(track_df,list_date)
```
