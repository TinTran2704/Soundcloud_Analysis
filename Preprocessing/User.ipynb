Tìm cột bị thiếu

```python
def Find_miss_col(df):
    l_col = []
    for item in df.columns:
        check = df[item].isnull()
        if check.sum() > 0:
            l_col.append(item)
    return l_col
Find_miss_col(user_df)
```

```python
def DelMissCol(df,percent):
    list_col = df.columns
    for col in list_col:
        check = (df[col].isnull()) #Tạo list check để lưu các giá trị null của cột(True là Null, False là khác Null)
        if(check.sum() > len(df)*percent): #Kiểm tra số giá trị rỗng có hơn số phần trăm cho trước
            del df[col]
    return df
user_df = DelMissCol(user_df,0.8)
```

```python
def DelMissRow(df,percent):
    dft = df.transpose()  #Chuyển vị để dễ thao tác với hàng hơn
    del_lis = []
    for item in dft.columns:
        check = (dft[item].isnull()) #Tạo list check để lưu các giá trị null của cột(True là Null, False là khác Null)
        if(check.sum() > len(dft)*percent): #Kiểm tra số giá trị rỗng có hơn số phần trăm cho trước
            del_lis.append(int(item))
    df = df.drop(df.index[del_lis],0)
    return df
user_df = DelMissRow(user_df,0.5)
```

Kiểu dữ liệu không phù hợp là created_at, last_modified, release_date, published_at, display_date. Chúng nên là Datetime.

```python
print(user_df['created_at'][0])
print(user_df['last_modified'][0])
list_date = ['created_at','last_modified']
```

```python
def Format_day(df, list_date):
    format_ = '%Y-%m-%dT%H:%M:%SZ'
    n = len(df['created_at'])
    for item in list_date:
        check = df[item].isnull()
        for i in range(n):
            if check[i] == False:
                df[item][i] = datetime.datetime.strptime(df[item][i],format_)
    return df
user_df = Format_day(user_df,list_date)
```
