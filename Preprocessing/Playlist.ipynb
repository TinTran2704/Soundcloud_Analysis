```python
def Find_miss_col(df):
    l_col = []
    for item in df.columns:
        check = df[item].isnull()
        if check.sum() > 0:
            l_col.append(item)
    return l_col
Find_miss_col(playlist_df)
```

```python
def DelMissCol(df,percent):
    list_col = df.columns
    for col in list_col:
        check = (df[col].isnull()) #T·∫°o list check ƒë·ªÉ l∆∞u c√°c gi√° tr·ªã null c·ªßa c·ªôt(True l√† Null, False l√† kh√°c Null)
        if(check.sum() > len(df)*percent): #Ki·ªÉm tra s·ªë gi√° tr·ªã r·ªóng c√≥ h∆°n s·ªë ph·∫ßn trƒÉm cho tr∆∞·ªõc
            del df[col]
    return df
playlist_df = DelMissCol(playlist_df,0.5)
```

```python
def DelMissRow(df,percent):
    dft = df.transpose()  #Chuy·ªÉn v·ªã ƒë·ªÉ d·ªÖ thao t√°c v·ªõi h√†ng h∆°n
    del_lis = []
    for item in dft.columns:
        check = (dft[item].isnull()) #T·∫°o list check ƒë·ªÉ l∆∞u c√°c gi√° tr·ªã null c·ªßa c·ªôt(True l√† Null, False l√† kh√°c Null)
        if(check.sum() > len(dft)*percent): #Ki·ªÉm tra s·ªë gi√° tr·ªã r·ªóng c√≥ h∆°n s·ªë ph·∫ßn trƒÉm cho tr∆∞·ªõc
            del_lis.append(int(item))
    df = df.drop(df.index[del_lis],inplace = False)
    return df
playlist_df = DelMissRow(playlist_df,0.5)
```

Ki·ªÉu d·ªØ li·ªáu kh√¥ng ph√π h·ª£p l√† created_at, last_modified, release_date, published_at, display_date. Ch√∫ng n√™n l√† Datetime.

```python
print(playlist_df['created_at'][0])
print(playlist_df['last_modified'][0])
print(playlist_df['published_at'][0])
print(playlist_df['display_date'][0])
list_date = ['created_at','last_modified','display_date','published_at']
```

```python
def Format_day(df, list_date):
    format_ = '%Y-%m-%dT%H:%M:%SZ'
    n = len(df['created_at'])
    for item in list_date:
        check = df[item].isnull()
        for i in range(n):
            if check[i] == False:
                df[item][i] = datetime.datetime.strptime(df[item][i],format_)
    return df
playlist_df = Format_day(playlist_df,list_date)
```
